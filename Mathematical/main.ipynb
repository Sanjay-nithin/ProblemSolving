## 1. Arithmetic Progression

### Required
- Common difference between adjacent arr elements
- first element (a₁ or a)
- last element (aₙ or l)
- size of the array (n)

## Formula
- To find the kᵗʰ term in the series
    ```bash
    a₁ is the first term
    a₂ = a₁ + d is the second term
    a₃ = a₁ + 2d is the third term
    
    So the kᵗʰ element will be:
    aₖ = a₁ + (k-1)d
    
    Or simply:
    aₖ = a + (k-1)d
    ```

- To find the sum of the AP series
    ```bash
    arr = [3, 7, 11, 15, 19, 23]
    
    sum of a₁ and aₙ is 3 + 23 = 26
    sum of a₂ and aₙ₋₁ is 7 + 19 = 26
    
    The sum of every pair of first and last element is the same
    There are total n/2 pairs to calculate first and last sum
    
    Sum formula:
    Sₙ = (n/2) × (a₁ + aₙ)
    or
    Sₙ = (n/2) × (a + l)
    
    If n is odd:
    Sₙ = (n/2) × (a + l) + a₍ₙ/₂₎ (mid value needs to be added)
    ```

````python
arr =[3,7,11,15,19,23]
a = 1
k = 5
d = 1
l = 10
n = 10
# print the 5th element
print(a+(k-1)*d)
# print the sum of arr elements
if n % 2 == 0:
    print((n//2)*(a+l))
else:
    print((n//2)*(a+l) + arr[n//2])
````

## 2. Geometric Progression

### Required
- Each term is multiplied by a constant ratio r:
    ```py
    a, a.r, a.r^2, a.r^3, ...
    ```
- first term (a)
- last term (l)
- common ratio (r)

### Formula
- To derive the nth term
    ```py
    an = ar^n-1
    ```
- To find the sum of n terms
    ```py
    a(l-r**n)//l-r
    ```


````python
arr = [1,2,4,8,16]
a = 1
l = 16
r = 2 
n = 5
# print kth element
k = 6
print((a*r)**(k-1))

# print sum
if r == 1:
    print(a * n)
else:
    print(a * (r**n - 1) // (r - 1))
````
