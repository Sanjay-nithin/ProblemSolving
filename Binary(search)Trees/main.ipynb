## Linked List implementation of Binary Tree

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = self.right = None 
    
class BinaryTree:
    def __init__(self):
        self.curr = None 
    
    def addNode(self, val, side=None, parent=None):
        if self.curr is None:
            self.curr = Node(val)
            return
        if side not in [1,2]:
            print("Invalid Side position value")
            exit()
            
        stack = [self.curr]
        while stack:
            node = stack.pop()
            
            if node.val == parent:
                if side == 1:
                    if node.left:
                        print("Element already exists in requested position")
                        exit()
                    else:
                        node.left = Node(val)
                        return
                elif side == 2:
                    if node.right:
                        print("Element already exists in requested position")
                        exit()
                    else:
                        node.right = Node(val)
                        return
            if node.left:
                stack.append(node.left)
            if node.right:
                stack.append(node.right)
        print("Parent value does not exist in binary Tree")
        exit()
    
    def display(self):
        curr = self.root
        queue = [curr]
        while queue:
            node = queue.pop(0)
            print(node.val, end=" ")
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        
        
bt = BinaryTree()
n = int(input("Number of nodes: "))
if n == 0:
    exit()
root = int(input("Enter the root element: "))
bt.addNode(root)
for i in range(1, n):
    side, parent, val = map(int, input("Side Parent Value: ").split())
    bt.addNode(val, side, parent)
bt.display()
```

## Array implementation of Binary Search Tree

```python
class ArrayBst:
    def __init__(self):
        self.arr = [] 
    
    def _extendArray(self, i):
        self.arr.extend([None] * (i-len(self.arr)+1))
    
    def addNode(self, val):
        if not self.arr:
            self.arr.append(val)
            return
        idx = 0
        while True:
            if self.arr[idx] is None:
                self.arr[idx] = val
                return 

            if val < self.arr[idx]:
                idx = 2*idx + 1
            elif val > self.arr[idx]:
                idx = 2*idx + 2
            self._extendArray(idx)
    
    def inorder(self):
        curr = self.arr 
        stack = []
        idx = 0
        while stack or idx < len(curr):
            while idx < len(curr) and curr[idx] is not None:
                stack.append(idx)
                idx = 2 * idx + 1
            idx = stack.pop()
            print(curr[idx], end=" ")
            idx = 2 * idx + 2
        

arr = ArrayBst()
for i in [20, 10, 30, 5, 9, 21, 31]:
    arr.addNode(i)
print(arr.inorder())
print(arr.arr)
```

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = self.right = None
    
class BST:
    def __init__(self):
        self.root = None

    def addNode(self, val):
        if self.root is None:
            self.root = Node(val)
            return 
        curr = self.root
        while True:
            if val < curr.val:
                if curr.left:
                    curr = curr.left
                else:
                    curr.left = Node(val)
                    return 
            elif val > curr.val:
                if curr.right:
                    curr = curr.right 
                else:
                    curr.right = Node(val) 
                    return
    def countNodes(self):
        def recursion(root):
            if root is None:
                return 0
            left = recursion(root.left)
            right = recursion(root.right)

            return left+right+1
        return recursion(self.root)



bt = BST()
for i in [20, 10, 30, 5, 11, 21, 31, 1]:
    bt.addNode(i)
print(bt.countNodes())
```

```python
def isSymmetric(root):
    if root is None:
        return True
    p, q = root.left, root.right

    def inorderCheck(p, q):
        if p is None and q is None:
            return True
        if p and q and p.val == q.val:
            return inorderCheck(p.left, q.right) and inorderCheck(p.right, q.left)
        return False
    return inorderCheck(p, q)



print(isSymmetric(root))

```

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None 

def display(root):
    if root is None:
        return 
    print(root.val, end="->")
    display(root.next)

def reverse(root):
    if root is None or root.next is None:
        return root
    new = reverse(root.next)
    root.next.next = root
    root.next = None
    return new 

root = Node(1)
root.next = Node(2)
root.next.next = Node(3)
root.next.next.next = Node(4)
root.next.next.next.next = Node(5)
display(root)
root = reverse(root)
print()
display(root)
```

```python
def printN(num, n):
    if num > n:
        return 
    print(num, end=" ")
    printN(num+1, n)

n = 10
printN(0, n)
```

```python
def reverse(arr, l, r):
    if l > r:
        return 
    arr[l], arr[r] = arr[r], arr[l]
    reverse(arr, l+1, r-1)
    
arr = [1,2,3,4,5]
reverse(arr, 0, 4)
print(arr)
```
